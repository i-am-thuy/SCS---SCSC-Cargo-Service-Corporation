##TỔNG QUAN VỀ SCS - CTCP Dịch vụ Hàng hóa Sài Gòn



Công ty chuyên cung cấp dịch vụ khai thác hàng hóa hàng không tại Cảng hàng không quốc tế Tân Sơn Nhất. 

Nhà ga SCS là nhà ga duy nhất của Việt Nam đạt tiêu chuẩn quốc tế của IATA (Hiệp hội vận chuyển hàng hóa quốc tế).
Khách hàng của SCS là các hãng hàng không lớn, có uy tín trên thế giới như Cathay Pacific, Singapore Airline, Emirates, … 

 * Hoạt động chính là kinh doanh, phát triển nhà ga hàng hóa tại sân bay Tân Sơn Nhất:
+ Khu vực sân đậu: 52,421 m2 có sức chứa máy bay 321 cùng lúc.
+ Khu vực ga hàng hóa: 26,670 m2 với công suất thiết kế 200,000 tấn hàng hóa (giai đoạn 1) và nâng lên 350,000 tấn hàng hóa (giai đoạn 2).
+ Khu vực nhà kho, văn phòng, công trình phụ trợ: 64,000 m2.

##Kết quả kinh doanh Quý I/2022:

+	Quý 1/2022, lợi nhuận Q1 2022 tăng mạnh do thị trường vận chuyển sôi động. Trong Q1/22, tổng sản lượng hàng hóa tăng 18.47% nhờ sản lượng hàng hóa quốc tế tăng mạnh (+40.77% svck) do hoạt động xuất khẩu sôi động, trong khi đó sản lượng hàng hóa quốc nội giảm 34.64% do nhà ga đã đạt công suất tối đa và SCS phải giới hạn công suất phục vụ hàng hóa nội địa. Giá bán trung bình tăng 6.9% svck do SCS áp dụng chính sách phí mới kể từ tháng 5/2021.

+	Kết quả kinh doanh Q1/22 tăng 25.4% svck đạt 246 tỷ đồng. Biên lợi nhuận gộp tăng 4,2 điểm % lên 83,5% giúp lợi nhuận gộp tăng 32,0% svck. Ngoài ra, doanh thu tài chính tăng 75,3% svck nhờ lượng tiền tích lũy gửi kỳ hạn ngắn tăng. Tổng hợp lại, LNR quý 1/22 tăng 36,6% svck đạt 188 tỷ đồng.

##LUẬN ĐỂM ĐẦU TƯ:

+	Về ngắn hạn, động lực tăng trưởng nhờ:
(1) Trong năm 2021, công ty dự kiến bắt đầu xây dựng tòa nhà văn phòng thứ hai, và có thể hoàn thành trong năm 2022. Tòa nhà mới nằm trên lô đất hiện tại, do đó không cần tính thêm chi đất vào chi phí hoạt động. Tổng diện tích cho thuê là 12.000 m2, với chi phí vốn dự kiến là 10 triệu USD, được tài trợ chủ yếu từ lợi nhuận giữ lại, do đó không cần vay ngân hàng quá nhiều. 
(2) Nhà ga của SCS đang hoạt động ở mức 133% công suất, do đó, công ty sẽ bắt đầu mở rộng công suất nhà ga từ 200.000 tấn/năm lên 350.000 tấn/năm trong năm 2022.

+	Về dài hạn, SCS được kỳ vọng là lựa chọn đầu tư an toàn do (1) công ty hoạt động trong thị trường độc quyền chỉ có hai công ty đang phát triển (SCS và TCS Công ty TNHH Dịch vụ Hàng hóa Tân Sơn Nhất). (2) Tình hình tài chính mạnh, không có dư nợ vay. (3) công ty đang thực hiện dự án M&A với các công ty vận tải hàng hóa ở Sân bay Nội Bài và dự án đầu tư nhà ga cảng hàng hóa ở sân bay Long Thành (dự kiến hoạt động vào năm 2025). 


##RỦI RO ĐẦU TƯ:

+	Rủi ro giảm giá do tình hình đại dịch/địa chính trị làm ảnh hưởng giao thông đường hàng không.
+	Các dự án tòa nhà văn phòng của SCS ở Tân Sơn Nhất và thương vụ M&A cần chờ thêm thông tin từ Đại hội Đảng. 
+	Dự án Long Thành đang trong quá trình thiết kế và SCS đang cung cấp cho Tổng Công ty Cảng Hàng không Việt Nam (ACV) kinh nghiệm về phương án bố trí sắp xếp càng hàng hóa hàng không. Tuy nhiên, Công ty chưa chắc chắn có thể giành được dự án Long Thành, đây là mối quan tâm chính cho các Nhà đầu tư khi quyết định đầu tư dài hạn vào SCS.
